This code is for matching flm and polscope image.

Resolution of FLM is slightly higher than that of polscope.

After matching, MFA of each cell can be freely accessible.

```python
import numpy as np
import matplotlib.pyplot as plt
import os, sys
import cv2
import copy
from common.Stitcher import Stitcher
from skimage.filters import threshold_otsu, threshold_triangle, threshold_yen, threshold_mean, threshold_minimum, threshold_li, threshold_isodata
from skimage.util import img_as_ubyte
from tqdm import tqdm
```

```python
#set resolution
FLM_x10=215.433 #nm/pixel
Polscope_x40=407.116/2 #nm/pixel

Scaling_factor=FLM_x10/Polscope_x40
```

Load all results

```python
current_path=os.getcwd()
npz_path=os.path.join(current_path, "npz_file")
save_path=os.path.join(current_path, "fig_save")

#set path
npz_files=os.listdir(npz_path)
npz_file_flm=[i for i in npz_files if "flm" in i]
npz_file_ret=[i for i in npz_files if "ret_angle_MFA_im_x40" in i]
npz_file_GT=[i for i in npz_files if "GT" in i]

#load data
#FLM
npz_flm=np.load(os.path.join(npz_path, npz_file_flm[0]), allow_pickle=True)
flm_b=npz_flm["im_stitch_b"]

#retardation
npz_ret=np.load(os.path.join(npz_path, npz_file_ret[0]), allow_pickle=True)
ret_im=npz_ret["ret_im"]
angle_im=npz_ret["angle_im"]
MFA_im=npz_ret["MFA_mat"]

#GT
npz_GT=np.load(os.path.join(npz_path, npz_file_GT[0]), allow_pickle=True)
nuclei=npz_GT["nuclei"]
boundary_im=npz_GT["boundary_image"]
```

```python
#resize FLM image for stitching
height, width=flm_b.shape

flm_b_resize=cv2.resize(flm_b, dsize=None, fx=Scaling_factor, fy=Scaling_factor)
boundary_im_resize=cv2.resize(boundary_im, dsize=None, fx=Scaling_factor, fy=Scaling_factor)
boundary_im_resize=np.uint8(np.where(boundary_im_resize>0, 255, 0))
nuclei_resize=cv2.resize(nuclei, dsize=None, fx=Scaling_factor, fy=Scaling_factor, interpolation=cv2.INTER_NEAREST)


#binarization of flm and ret im
threshold = threshold_triangle(ret_im)
ret_binary=np.where(ret_im<threshold, 255, 0)
threshold = threshold_otsu(flm_b_resize)
flm_b_binary=np.where(flm_b_resize<threshold, 255, 0)

fig, ax=plt.subplots(2, 2, figsize=(12, 8))
ax[0, 0].imshow(ret_binary)
ax[0, 1].imshow(flm_b_binary)
ax[1, 0].imshow(boundary_im_resize)
ax[1, 1].imshow(nuclei_resize, cmap="jet")
plt.show()
plt.close()
```

```python
#matching ret and flm im
stitcher=Stitcher()
ret_im_adj, flm_b_im_adj=stitcher.image_dimension_adj_const(ret_binary, flm_b_binary, padding=255, const=255, criterion="A")
ret_im_adj, flm_b_im_ori_adj=stitcher.image_dimension_adj_const(ret_binary, flm_b_resize, padding=0, const=255, criterion="A")
ret_im_adj, boundary_im_adj=stitcher.image_dimension_adj(ret_binary, boundary_im_resize, padding=0, criterion="A")
ret_im_adj, nuclei_im_adj=stitcher.image_dimension_adj(ret_binary, nuclei_resize, padding=0, criterion="A")

#image matching of ret and flm
_, delete_line_h, delete_line_v, H=stitcher.stitch((np.uint8(ret_im_adj), np.uint8(flm_b_im_adj)))

#image matching of others under the same condition
flm_b_matching=cv2.warpPerspective(flm_b_im_adj, H, (flm_b_im_adj.shape[1], flm_b_im_adj.shape[0]), borderValue=255)
flm_b_ori_matching=cv2.warpPerspective(flm_b_im_ori_adj, H, (flm_b_im_ori_adj.shape[1], flm_b_im_ori_adj.shape[0]))
boundary_matching=cv2.warpPerspective(boundary_im_adj, H, (boundary_im_adj.shape[1], boundary_im_adj.shape[0]))
nuclei_matching=cv2.warpPerspective(nuclei_im_adj, H, (nuclei_im_adj.shape[1], nuclei_im_adj.shape[0]), flags=cv2.INTER_NEAREST)
```

```python
fig, ax=plt.subplots(2, 2, figsize=(12, 8))
ax[0, 0].imshow(ret_im_adj)
ax[0, 1].imshow(flm_b_matching)
ax[1, 0].imshow(boundary_matching)
ax[1, 1].imshow(nuclei_matching, cmap="jet")
#plt.savefig(os.path.join(save_path, "Image_matching_flm_ret.png"), dpi=600)
plt.show()
plt.close()
```

```python
#delete background as long as possible
delete_line_h = np.where(np.all(flm_b_matching == 255, axis=0))[0]###全ての値が０の列番号を抽出
delete_line_v = np.where(np.all(flm_b_matching == 255, axis=1))[0]###全ての値が０の列番号を抽出
#FLM binary
flm_b_trim = np.delete(flm_b_matching, delete_line_h, axis=1)
flm_b_trim = np.delete(flm_b_trim, delete_line_v, axis=0)
#FLM ori
flm_b_ori_trim = np.delete(flm_b_ori_matching, delete_line_h, axis=1)
flm_b_ori_trim = np.delete(flm_b_ori_trim, delete_line_v, axis=0)
#ret
ret_trim = np.delete(ret_im, delete_line_h, axis=1)
ret_trim = np.delete(ret_trim, delete_line_v, axis=0)
#angle
angle_trim = np.delete(angle_im, delete_line_h, axis=1)
angle_trim = np.delete(angle_trim, delete_line_v, axis=0)
#boundary
boundary_trim = np.delete(boundary_matching, delete_line_h, axis=1)
boundary_trim = np.delete(boundary_trim, delete_line_v, axis=0)
#nulcei
nuclei_trim = np.delete(nuclei_matching, delete_line_h, axis=1)
nuclei_trim = np.delete(nuclei_trim, delete_line_v, axis=0)
#MFA
MFA_trim = np.delete(MFA_im, delete_line_h, axis=1)
MFA_trim = np.delete(MFA_trim, delete_line_v, axis=0)

#binarization
#FLM
threshold = threshold_otsu(flm_b_trim)
flm_b_trim_binary=np.where(flm_b_trim>threshold, 255, 0)
#boundary
threshold = threshold_otsu(boundary_trim)
boundary_trim_binary=np.where(boundary_trim>threshold, 255, 0)
```

```python
#labels in original nuclei image which are already omitted in trimmed nuclei image are set to 0
labels_trim_unique=np.unique(nuclei_trim.flatten())
height, width=nuclei.shape
labels_omit=list(set(np.unique(nuclei.flatten()))-set(labels_trim_unique))
nuclei_copy=copy.deepcopy(nuclei).flatten()

for label_omit in tqdm(labels_omit):
    label_index=np.where(nuclei_copy==label_omit)
    nuclei_copy[label_index[0]]=0

nuclei_copy=nuclei_copy.reshape((nuclei.shape[0], nuclei.shape[1]))
```

```python
#save results
np.savez(os.path.join(npz_path, "FLM_ret_matching_results_x40.npz"), 
        ret_im=ret_trim, angle_im=angle_trim, MFA_im=MFA_trim, flm_b_binary=flm_b_trim_binary, nuclei_im=nuclei_trim, nuclei_ori=nuclei_copy,
        boundary_im=boundary_trim_binary, boundary_im_ori=boundary_im, flm_b=flm_b_ori_trim, flm_b_ori=flm_b)
```

```python
#show matching reslt by overlaying boundary image on ret image
ret_trim_copy=copy.deepcopy(ret_trim)
boundary_im_expand=cv2.GaussianBlur(np.uint8(boundary_trim_binary),(7,7),0)
boundary_im_expand=np.where(boundary_im_expand>0, 255, 0)
boundary_index=np.where(boundary_trim_binary==255)

ret_trim_copy[boundary_index]=255

plt.imshow(ret_trim_copy, cmap="jet")
plt.savefig(os.path.join(save_path, "FLM_ret_matching_boundary_overlay_x40.png"), dpi=1500)
plt.show()
plt.close()
```
