This code is for segmentation of FLM to avoid undersegmentation around ray and resin cell region.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os, sys
import cv2
import copy
import mahotas as mh
from skimage.measure import label
from tqdm import tqdm
from skimage.filters import threshold_otsu, threshold_triangle, threshold_yen, threshold_mean, threshold_minimum, threshold_li, threshold_isodata
import itertools
from multiprocessing import Pool
import multiprocessing
import common.Watershed_func
from matplotlib import patches
```

```python
#set dict object
threshold_dict={"otsu": threshold_otsu,
                "triangle":threshold_triangle,
                "li":threshold_li,
                "mean":threshold_mean,
                "yen":threshold_yen,
                "minimum":threshold_minimum,
                "isodata":threshold_isodata}
threshold_list=["otsu", "triangle", "li", "mean", "yen", "minimum", "isodata"]
```

```python
current_path=os.getcwd()

save_path=os.path.join(current_path, "fig_save")
if os.path.exists(save_path)==False:
    os.makedirs(save_path)
    
npz_path=os.path.join(current_path, "npz_file")
if os.path.exists(npz_path)==False:
    os.makedirs(npz_path)
```

```python
#import x10 images
npz_file=os.listdir(npz_path)
npz_file=[i for i in npz_file if "flm" in i]
npz_file_path=os.path.join(npz_path, npz_file[0])
npz=np.load(npz_file_path)
im_stitch_b=npz["im_stitch_b"]
im_stitch_r=npz["im_stitch_r"]
im_stitch_UV=npz["im_stitch_ori_UV"]
im_stitch=npz["im_stitch_binary"]
```

```python
#binarization
#UV excitation
threshold = threshold_dict["otsu"](im_stitch_b)
im_binary_b=np.where(im_stitch_b<threshold, 255, 0)

#Green excitatino
threshold = threshold_dict["otsu"](im_stitch_r)
im_binary_r=np.where(im_stitch_r<threshold, 0, 255)
```

```python
#omit small noise inside lumens using connected component 
watershed_preprocessing=Watershed_func.watershed_preprocessing()

labels_lumen=watershed_preprocessing.omit_by_area(im_binary_b, area=3000)
original_bin=watershed_preprocessing.omit_by_area_aspect_ratio(labels_lumen, area_1=100, area_2=25, aspect_ratio=1/1.5)

fig, ax=plt.subplots(1, 2, figsize=(10, 4))
ax[0].imshow(labels_lumen)
ax[1].imshow(original_bin)
plt.show()
```

Ray correction

```python
labels_ray=watershed_preprocessing.ray_extraction(im_binary_r, kernel=(5, 5), iterations=3, area=100, aspect_ratio=0.4)
#overlay ray region on binarization image
original_bin_copy=copy.deepcopy(original_bin)
original_bin_copy[labels_ray==255]=255

fig, ax=plt.subplots(1, 2, figsize=(10, 4))
ax[0].imshow(labels_ray)
ax[1].imshow(original_bin_copy)
plt.show()
```

Watershed segmentation applied to with and with ray region correction

```python
#set seed
watershed=Watershed_func.watershed()
nuclei, lines=watershed.watershed_segmentation(original_bin_copy)

plt.imshow(nuclei[:, :], cmap="gist_rainbow")
plt.colorbar()
#plt.savefig(os.path.join(save_path, str(os.path.splitext(image_path[i])[0])+"_labeled_image.png"), dpi=600)
plt.show()
```

```python
#create boundary image with good visualization
boundary_image= cv2.GaussianBlur(np.uint8(lines*255), (3,3), 0)
boundary_image = np.where(boundary_image[:,:]>0, 255, 0)
```

```python
core_num=3
watershed_postprocessing=Watershed_func.watershed_postprocessing(nuclei=nuclei, core_num=core_num)
im_overlay=watershed_postprocessing.boundary_overlay(im_stitch_UV, np.uint8(boundary_image), (3, 3))

plt.figure(figsize=(6,6))
plt.imshow(im_overlay)
plt.savefig(os.path.join(save_path, "wateshed_result_flm.png"), dpi=600)
plt.show()
```

```python
#Original image is too large for manipulation
#As temporary solution,image is divided into 4 part
im_stitch_UV=cv2.cvtColor(im_stitch_UV, cv2.COLOR_RGB2BGR)

height, width=boundary_image.shape
height_div=2
width_div=2

for i in range(height_div):
    for j in range(width_div):
        boundary_image_div=boundary_image[i*(height//height_div):(i+1)*(height//height_div), j*(width//width_div):(j+1)*(width//width_div)]
        im_stitch_UV_div=im_stitch_UV[i*(height//height_div):(i+1)*(height//height_div), j*(width//width_div):(j+1)*(width//width_div), :]
        
        fig, ax=plt.subplots(1,2 , figsize=(10, 4))
        ax[0].imshow(boundary_image_div)
        ax[1].imshow(im_stitch_UV_div)
        plt.show()
        
        #save im zero for manual correction
        watershed_postprocessing.save_overlay_im(boundary_image_div, im_stitch_UV_div, save_path, str(i)+"_"+str(j))
```
