The purpose of this code is to extract geometrical parameters of wood cell components utilizing segmnetation boundaries determined by watershed segmentation (code 1).

Target paremetes are below

1) cell and lumen area

2) cell and lumen perimeter

3) cell and lumen circularity

4) center of gravity of cell and lumen

5) cell wall occupancy


```python
import matplotlib.pyplot as plt
import numpy as np
import os, sys
import cv2
import copy
import mahotas as mh
from skimage.filters import threshold_otsu
from tqdm import tqdm
import common.Stitcher
```

```python
#set resolution
FLM_x10=215.433 #nm/pixel
Polscope_x40=407.116/2 #nm/pixel

Scaling_factor=FLM_x10/Polscope_x40
```

```python
#set path
current_path=os.getcwd()
npz_path=os.path.join(current_path, "npz_file")
npz_files=os.listdir(npz_path)
npz_file=[i for i in npz_files if "FLM_ret_matching_results_x40" in i]

#load npz
npz=np.load(os.path.join(npz_path, npz_file[0]))
flm=npz["flm_b_ori"]
flm_adj=npz["flm_b"]
boundary_im=npz["boundary_im_ori"]
boundary_im_adj=npz["boundary_im"]
nuclei=npz["nuclei_ori"]
nuclei_trim=npz["nuclei_im"]

#binarization
threshold = threshold_otsu(flm)
flm_binary=np.where(flm>threshold, 0, 255)
```

```python
#remove the cell components touching the frame of the micrograph
nuclei_inner = mh.labeled.remove_bordering(nuclei)
#relabeled_nuclei, n_left = mh.labeled.relabel(nuclei_inner)

fig, ax=plt.subplots(1, 3, figsize=(14, 4))
ax[0].imshow(nuclei)
ax[1].imshow(nuclei_inner)
ax[2].imshow(flm_binary)
#plt.title(str(name_list[id])+"\nafter removing cell components touching the frame of the micrograph")
plt.show()
plt.close()
```

```python
omit_labels=list(set(np.unique(nuclei_trim.flatten()))-set(np.unique(nuclei_inner.flatten())))
nuclei_trim_copy=copy.deepcopy(nuclei_trim).flatten()

for omit_label in tqdm(omit_labels):
    index=np.where(nuclei_trim.flatten()==omit_label)
    nuclei_trim_copy[index]=0
```

```python
plt.imshow(nuclei_trim_copy.reshape((nuclei_trim.shape[0], nuclei_trim.shape[1])), cmap="jet")
plt.show()
plt.close()
```

```python
#relabelling nuclei image
label_array=np.arange(0, len(np.unique(nuclei_inner)), dtype=np.int16)
label_sort_array=np.sort(np.unique(nuclei_inner))

nuclei_inner=nuclei_inner.flatten()
nuclei_trim_copy=nuclei_trim_copy.flatten()

relabeled_nuclei=np.zeros((len(nuclei_inner)), dtype=np.int16)
relabeled_nuclei_adj=np.zeros((len(nuclei_trim_copy)), dtype=np.int16)

for i in tqdm(range(len(label_array))):
    index=np.where(nuclei_inner==label_sort_array[i])
    index_adj=np.where(nuclei_trim_copy==label_sort_array[i])
    
    relabeled_nuclei[index]=int(label_array[i])
    relabeled_nuclei_adj[index_adj]=int(label_array[i])
    
relabeled_nuclei=relabeled_nuclei.reshape((nuclei.shape[0], nuclei.shape[1]))
relabeled_nuclei_adj=relabeled_nuclei_adj.reshape((nuclei_trim.shape[0], nuclei_trim.shape[1]))
```

```python
#show result
fig, ax=plt.subplots(1,2, figsize=(12, 4))
ax[0].imshow(relabeled_nuclei)
ax[1].imshow(relabeled_nuclei_adj)
plt.show()
plt.close()
```

Anatomical parameter extraction

```python
#set lists for each parameters
cell_area_list=[] #cell area
cell_perimeter_list=[]
cell_cx_list=[] #y coordinate of the center of the gravity
cell_cy_list=[] #x cooridinate of the center of the gravity
cell_circle_level_list=[] #circurality

lumen_area_list=[] #cell area
lumen_perimeter_list=[]
lumen_cx_list=[] #y coordinate of the center of the gravity
lumen_cy_list=[] #x cooridinate of the center of the gravity
lumen_circle_level_list=[] #circurality


label_array=np.unique(relabeled_nuclei)
for i in tqdm(label_array):
    #############################################
    # anatomical parameter based on cell region #
    #############################################
    
    #extract cell area labelled i
    cell_ex=np.where(relabeled_nuclei==i, 255, 0)
    
    
    #detecting contours around cell boundary
    cell_contours, cell_hierarchy = cv2.findContours(np.uint8(cell_ex), cv2.RETR_TREE,  cv2.CHAIN_APPROX_NONE)
    cell_contours_len_list=[len(cell_contours[j]) for j in range(len(cell_contours))]
    cell_ind_max=np.argmax(cell_contours_len_list)
    cell_contours=cell_contours[cell_ind_max]

    #cell area
    cell_area = cv2.contourArea(cell_contours)
    #cell perimeter length
    cell_perimeter = cv2.arcLength(cell_contours,True)

    #center of the gravity is calculated based on contour
    cell_mu = cv2.moments(cell_contours, False)
    if cell_mu["m00"]==0:
        cell_x=0
        cell_y=0
    else:
        cell_x, cell_y= int(cell_mu["m10"]/cell_mu["m00"]) , int(cell_mu["m01"]/cell_mu["m00"])

    #cell circular level
    cell_circle_level = 4.0*np.pi*cell_area/(cell_perimeter*cell_perimeter)
    
    
    
    ##############################################
    # anatomical parameter based on lumen region #
    ##############################################
    
    #all cell components consist of 3 part, bg, cell wall and lumen
    lumen_ex=np.where(relabeled_nuclei==i, flm_binary, 0)
    #lumen_ex=cv2.bitwise_not(lumen_ex)
    #plt.imshow(lumen_ex)
    #plt.show()
    
    if np.count_nonzero(lumen_ex==255)>0:
    
        #In FLM binary image, 0 is assigned to lumen regions 
        #extract cell lumen contour
        lumen_contours, lumen_hierarchy = cv2.findContours(np.uint8(lumen_ex), cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
        lumen_contours_len_list=[len(lumen_contours[j]) for j in range(len(lumen_contours))]
        if len(lumen_contours_len_list)>0:
            lumen_ind_max=np.asarray(lumen_contours_len_list).argsort()[::-1][0]
        else:
            pass

        lumen_contours = lumen_contours[lumen_ind_max]
    
        #lumen area
        lumen_area = cv2.contourArea(lumen_contours)
    
        #lumen perimeter length
        lumen_perimeter = cv2.arcLength(lumen_contours,True)
    
        #center of the gravity is calculated based on contour
        lumen_mu = cv2.moments(lumen_contours,  False)
        if lumen_mu["m00"]==0:
            lumen_x=cell_x
            lumen_y=cell_y
        else:
            lumen_x, lumen_y= int(lumen_mu["m10"]/lumen_mu["m00"]) , int(lumen_mu["m01"]/lumen_mu["m00"])

        #cell circular level
        if lumen_perimeter==0:
            lumen_circle_level=0
        else:
            lumen_circle_level = 4.0*np.pi*lumen_area/(lumen_perimeter*lumen_perimeter)
        
    else:
        lumen_area=0
        lumen_perimeter=0
        lumen_circle_level=0
        lumen_x=cell_x
        lumen_y=cell_y
    
    ###########################
    # summarize all parameter #
    ###########################
    
    #cell
    cell_area_list.append(cell_area)
    cell_perimeter_list.append(cell_perimeter)
    cell_cx_list.append(cell_x)
    cell_cy_list.append(cell_y)
    cell_circle_level_list.append(cell_circle_level)
   
    
    #lumen
    lumen_area_list.append(lumen_area)
    lumen_perimeter_list.append(lumen_perimeter)
    lumen_cx_list.append(lumen_x)
    lumen_cy_list.append(lumen_y)
    lumen_circle_level_list.append(lumen_circle_level)
  

cell_area_list=np.asarray(cell_area_list)
cell_perimeter_list=np.asarray(cell_perimeter_list)
cell_cx_list=np.asarray(cell_cx_list)
cell_cy_list=np.asarray(cell_cy_list)
cell_circle_level_list=np.asarray(cell_circle_level_list)
lumen_area_list=np.asarray(lumen_area_list)
lumen_perimeter_list=np.asarray(lumen_perimeter_list)
lumen_cx_list=np.asarray(lumen_cx_list)
lumen_cy_list=np.asarray(lumen_cy_list)
lumen_circle_level_list=np.asarray(lumen_circle_level_list)
```

```python
#calculate cell wall ocuupancy
cellwall_occupancy_list=[]
cellwall_area_list=[]
for i in tqdm(range(len(cell_area_list))):
    cellwall_occupancy_list.append((1-lumen_area_list[i]/cell_area_list[i])*100)
    cellwall_area_list.append(cell_area_list[i]-lumen_area_list[i])
```

```python
#save results
np.savez_compressed(os.path.join(npz_path, "anatomical_parameters_x40.npz"), 
                    cell_area_list=cell_area_list, cell_perimeter_list=cell_perimeter_list,
                    cell_cx_list=cell_cx_list, cell_cy_list=cell_cy_list,
                    cell_circle_level_list=cell_circle_level_list,
                    lumen_area_list=lumen_area_list, lumen_perimeter_list=lumen_perimeter_list, 
                    lumen_cx_list=lumen_cx_list, lumen_cy_list=lumen_cy_list, 
                    lumen_circle_level_list=lumen_circle_level_list,
                    cellwall_occuapncy_list=cellwall_occupancy_list,
                    cellwall_area_list=cellwall_area_list,
                    label_map=relabeled_nuclei, nuclei_trim=relabeled_nuclei_adj,
                    flm=flm)
```
